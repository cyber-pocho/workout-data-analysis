```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load
df = pd.read_csv('/content/drive/MyDrive/data/workout_activities.csv')

# Feature Engineering
df['CaloriesPerMinute'] = df['Calories'] / df['Duration']

# Summary Stats
print(df.columns)
print(df.groupby('WorkoutType')['CaloriesPerMinute'].mean().to_string())
print(df.groupby('WorkoutType')[['Pulse', 'MaxPulse']].mean().to_string())

# Visuals
plt.figure()
sns.barplot(data=df, x='WorkoutType', y='Calories').set_title('Calories per Workout Type')

plt.figure()
sns.barplot(data=df, x='WorkoutType', y='CaloriesPerMinute').set_title('Calories per Minute')

plt.figure()
sns.barplot(data=df, x='Duration', y='CaloriesPerMinute')
plt.title('Duration vs Calories per Minute')

plt.tight_layout()
plt.show()

# Export
df.to_csv('/content/drive/MyDrive/data/cleaned_workout_data.csv', index=False)
```
